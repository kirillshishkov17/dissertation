```python
# Строит график зависимости коэффициента наклона ротора от угла наклона

import numpy as np
import matplotlib.pyplot as plt

def calculate_C_beta(beta):
    return (
        1 - 0.185 
        + 0.285 * np.cos(0.105 * beta) + 0.014 * np.sin(0.105 * beta)
        - 0.144 * np.cos(0.209 * beta) - 0.017 * np.sin(0.209 * beta)
        + 0.06 * np.cos(0.314 * beta) + 0.011 * np.sin(0.314 * beta)
        - 0.016 * np.cos(0.419 * beta) - 0.005 * np.sin(0.419 * beta)
    )

# Создаем массив значений beta в градусах
beta = np.linspace(-15, 15, 300)

# Вычисляем значения функции
values = calculate_C_beta(beta)

plt.figure(figsize=(10, 6))
plt.plot(beta, values)
plt.xlabel('Угол наклона β (градусы)')
plt.ylabel('C_beta')
plt.title('График функции C_beta(β)')
plt.grid(True)
plt.show()
```

```python
# Строит график зависимости коэффициента рыскания в зависимости от угла рыскания

import numpy as np
import matplotlib.pyplot as plt

def yaw_coefficient(gamma_deg):
    gamma_rad = np.deg2rad(gamma_deg)
    return np.cos(gamma_rad) ** 3

# Диапазон углов рыскания в градусах
gamma_deg = np.linspace(-30, 30, 300)

# Вычисление коэффициента
c_gamma = yaw_coefficient(gamma_deg)

# Построение графика
plt.figure(figsize=(8, 5))
plt.plot(gamma_deg, c_gamma, color='royalblue', linewidth=2)
plt.xlabel('Угол рыскания γ (градусы)')
plt.ylabel('Коэффициент рыскания $C_\\gamma$')
plt.title('Зависимость коэффициента рыскания от угла рыскания')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
